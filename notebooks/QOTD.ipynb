### This notebook scrapes quotes on website into a dataframe and saves it into a csv. The dataset can be used to create a random quote generator. 

```python
from urllib.request import Request, urlopen
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
```

```python
site = "https://parade.com/937586/parade/life-quotes/"
hdr = {'User-Agent': 'Mozilla/5.0'}
bookpage = requests.get(site)
soup = BeautifulSoup(bookpage.text, "html.parser")
```

```python
quotes_db = []
quotes = soup.find_all('p')
for i in range(0,152):
    quote = quotes[i+8].text.strip()
    quotes_db.append(quote)
```

```python
#quotes_db
```

```python
df = pd.DataFrame(np.array(quotes_db))
```

```python
df.columns = ['quote']
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

```python
new = df["quote"].str.split(".", n = 1, expand = True) 
```
