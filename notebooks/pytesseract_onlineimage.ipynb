### This notebook describes how to extract text data from images hosted online.   
The example used here is about public places visited by COVID-19 cases while infectious. The list includes public places COVID-19 cases had visited for more than 30 minutes and is released by the Ministry of Health Singapore beginning 25 May. It is compiled by Channel News Asia into a table and posted as an image in an [article](https://www.channelnewsasia.com/news/singapore/covid-19-cases-ntuc-fairprice-bukit-panjang-plaza-hillion-mall-12776058). We want to extract the text data within the image.   

Make sure [pytesseract](https://pypi.org/project/pytesseract/) is installed. Instructions on installation can be found [here](https://www.simplifiedpython.net/how-to-extract-text-from-image-in-python/).

```python
import pytesseract
```

```python
try:
    from PIL import Image
except ImportError:
    import Image
```

```python
import requests
from io import BytesIO

url = 'https://www.channelnewsasia.com/image/12776060/0x0/816/512/6d8e49979a852f751285bb6e29793659/pn/covid-19-visit-public-places-may-27.png'
response = requests.get(url)
img = Image.open(BytesIO(response.content))
```

```python
img
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'
```

```python
print(pytesseract.image_to_string(img))
```

The extracted data is shared on: https://data.world/hxchua/public-places-visited-in-singapore-by-covid-19-cases.
